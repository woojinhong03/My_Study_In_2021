```python
import numpy as np
from scipy import sparse

eye = np.eye(8, 10)
print(eye,'\n \n')

s = sparse.csr_matrix(eye)
print(s)

```

```python
test = np.diag((4,2,2,5,1,6))
print(test,'\n \n')

s_test = sparse.csc_matrix(test)
print(s_test)
```

```python
import matplotlib.pyplot as plt

# -10 부터 10까지 100개의 간격으로 나뉜 배열 생성
x = np.linspace(-10, 10, 100)

# sin 함수를 사용하여 y 배열 생성
y = np.sin(x)

# plot 함수는 한 배열 값을 다른 배열에 대응해서 선 그래프를 그림
plt.plot(x, y, marker = "o")
```

```python
from IPython.display import display
import pandas as pd

# 회원 정보가 들어간 간단한 데이터셋 생성
data = {
        'Name' : ['john','anna','peter','linda'],
        'Location' : ['new york','paris','berlin','london'],
        'Age' : [24,13,53,33]
        }
data_pandas = pd.DataFrame(data)

#IPython.display 는 주피터 노트북에서 Dataframe 깔끔하게 출력
display(data_pandas)
```

```python
#age 열의 값이 30 이상인 행 선택
display(data_pandas[data_pandas.Age >= 30])
```
