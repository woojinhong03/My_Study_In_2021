```python
from sklearn.datasets import  load_breast_cancer
cancer = load_breast_cancer()
print("cancer.keys(): {}".format(cancer.keys()))
#데이터 포인트 개수 및 특성 수
print("data: {}".format(cancer.data.shape))
#569개의 class sample num
print("class sample num {}".format(
    {n:v for n, v in zip(cancer.target_names, np.bincount(cancer.target))}
))
```

```python
mglearn.plots.plot_knn_classification(n_neighbors=5)
```

```python
#train_test_split 임포트하여 데이터를 훈련 세트와 테스트 세트로 나눔
from sklearn.model_selection import train_test_split
X,y = mglearn.datasets.make_forge()
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=0)
#knn 임포트하고 객체 생성
from sklearn.neighbors import KNeighborsClassifier
#이웃의 수 매개변수 지정
clf = KNeighborsClassifier(n_neighbors=7)
#훈련 세트를 사용해서 분류 모델 학습
clf.fit(X_train, y_train)
#테스트 데이터를 사용해서 예측
print("테스트 세트 예측: ",clf.predict(X_test))
#모델이 얼마나 잘 일반화 되었는지 정확도 평가
print("정확도: {:.2f}".format(clf.score(X_test,y_test)))
```
